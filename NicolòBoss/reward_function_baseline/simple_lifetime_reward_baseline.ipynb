## Smoke test

This checks the initial knob unpacking and evaluates one full lifetime loss. It does not launch a full CMA-ES benchmark.

```python
x0 = jnp.array([1.0, 0.0, 4.0, 0.0])

g2_0, epsd_0 = unpack_knobs(x0)
print("g2:", g2_0)
print("eps_d:", epsd_0)

loss0 = cat_lifetime_loss(
    x0,
    eta_target=100.0,
    lambda_bias=2.0,
)

print("Initial cat lifetime loss:", loss0)
```

## Baseline metrics plot

Run the baseline once and plot the fitted loss, `Tx`, `Tz`, and `eta = Tz / Tx`.

```python
from pathlib import Path
from matplotlib import pyplot as plt

metrics = cat_lifetime_metrics(
    x0,
    eta_target=100.0,
    lambda_bias=2.0,
)

print("loss:", metrics["loss"])
print("Tx:", metrics["Tx"])
print("Tz:", metrics["Tz"])
print("eta:", metrics["eta"])

fig, axes = plt.subplots(1, 4, figsize=(12, 3))
for ax, key in zip(axes, ["loss", "Tx", "Tz", "eta"]):
    ax.bar([key], [metrics[key]], color="#2f6f73")
    ax.set_title(key)
    ax.grid(axis="y", alpha=0.25)

fig.tight_layout()
plot_path = Path("baseline_lifetime_metrics.png")
fig.savefig(plot_path, dpi=160, bbox_inches="tight")
plt.show()
print("Saved plot to", plot_path)
```
